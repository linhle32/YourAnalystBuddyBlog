Template for log transformation and boxcox transformation for handling skewed data. Codes explanation are at https://youranalystbuddy.com/handle-skewed-data/

## Loading and investigating distributions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('students-skewed.csv')

data.describe()
```

```python
data.hist(bins=10, figsize=(7,7))
plt.show()
```

## Log transformation

```python
data_log = np.log(data)

data_log.hist(bins=10, figsize=(7,7))
plt.show()
```

## Boxcox transformation

```python
from scipy.stats import boxcox

data_boxcox = data.copy()
for col in data:
    data_boxcox[col] = boxcox(data[col])[0]
    
data_boxcox.hist(bins=10, figsize=(8,8))
plt.show()
```
