## Loading data

```python
import pandas as pd

students = pd.read_csv('students1000.csv')
students = students.drop('StudentID', axis=1)
students.head(2)
```

## Side by side boxplots

```python
import matplotlib.pyplot as plt
```

```python
students[['Major','FirstYearGPA']].boxplot(by='Major', rot=20, figsize=(6,6))
plt.show()
```

## Stacked bar charts

```python
pd.crosstab(students['State'],students['Major']).plot.bar(stacked='True', rot=20, figsize=(6,6))
plt.show()
```

```python
crosstab = pd.crosstab(students['Major'],students['State'])
crosstab = crosstab / crosstab.sum() *100
crosstab.T.plot.bar(stacked='True',rot=20,figsize=(6,6))
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
