```python
import numpy as np
import matplotlib.pyplot as plt

x = np.sort(np.random.normal(0,1,(100,1)),axis=0)
y = 2*x + 0.3*x**2 + 0.5*x**3 + np.random.normal(0,2,(100,1))

plt.figure(figsize=(4,4))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x,y)
yh = lr.predict(x)
plt.figure(figsize=(4,4))
plt.scatter(x,y)
plt.plot(x,yh,c='red',linewidth=4)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree=15)
x15 = pf.fit_transform(x)
lr.fit(x15,y)
yh15 = lr.predict(x15)
```
