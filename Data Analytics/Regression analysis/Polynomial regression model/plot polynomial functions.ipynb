```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,400)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x, x, c='red', linewidth=3)
plt.scatter(x, x + np.random.normal(0,1,400))
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(x, x**2, c='red', linewidth=3)
plt.scatter(x, x**2 + np.random.normal(0,1,400))
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(x, x**3, c='red', linewidth=3)
plt.scatter(x, x**3 + np.random.normal(0,1,400))
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(x, x**4, c='red', linewidth=3)
plt.scatter(x, x**4 + np.random.normal(0,1,400))
plt.axis('off')
plt.show()
```

```python
x1 = np.random.normal(0,1,500)
x2 = np.random.normal(0,1,500)
y = 1 + 0.5*x1*x2 + np.random.normal(0,0.25,500)

plt.figure(figsize=(5,5))
plt.scatter(x1,y)
plt.axis('off')
plt.show()
plt.figure(figsize=(5,5))
plt.scatter(x2,y)
plt.axis('off')
plt.show()
plt.figure(figsize=(5,5))
plt.scatter(x1*x2,y)
plt.axis('off')
plt.show()
```
