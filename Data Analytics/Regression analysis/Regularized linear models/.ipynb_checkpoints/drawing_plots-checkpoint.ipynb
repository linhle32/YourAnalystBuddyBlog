```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,1000)
y1 = x**2
y2 = np.abs(x)
```

```python
plt.figure(figsize=(4,4))
plt.scatter([-1,0.98],[1,1],c='red',s=75,zorder=10)
plt.plot(x,y1,c='orange',linewidth=3,zorder=5)
plt.xlabel('a')
plt.ylabel('Penalty')
# plt.tick_params(
#     axis='both',
#     which='both',     
#     bottom=False,     
#     top=False,
#     left=False,
#     labelleft=False,
#     labelbottom=False)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(4,4))
plt.scatter([-1,0.98],[1,1],c='red',s=75,zorder=10)
plt.plot(x,y2,c='green',linewidth=3,zorder=5)
plt.ylim(0,4)
plt.xlabel('a')
plt.ylabel('Penalty')
# plt.tick_params(
#     axis='both',
#     which='both',     
#     bottom=False,     
#     top=False,
#     left=False,
#     labelleft=False,
#     labelbottom=False)
plt.show()
plt.clf()
```

```python
y3 = 0.7*y1 + 0.3*y2
plt.figure(figsize=(4,4))
plt.scatter([-1,0.98],[1,1],c='red',s=75,zorder=10)
plt.plot(x,y3,c='teal',linewidth=3,zorder=5)
plt.ylim(0,4)
plt.xlabel('a')
plt.ylabel('Penalty')
# plt.tick_params(
#     axis='both',
#     which='both',     
#     bottom=False,     
#     top=False,
#     left=False,
#     labelleft=False,
#     labelbottom=False)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(4,4))
plt.scatter([-1,0.98],[1,1],c='red',s=75,zorder=10)
plt.plot(x,y1,c='orange',linewidth=3,zorder=5)
plt.plot(x,y2,c='green',linewidth=3,zorder=3)
plt.plot(x,y3,c='teal',linewidth=3,zorder=0)
plt.xlabel('a')
plt.ylabel('Penalty')
plt.tick_params(
    axis='both',
    which='both',     
    bottom=False,     
    top=False,
    left=False,
    labelleft=False,
    labelbottom=False)
plt.show()
plt.clf()
```
