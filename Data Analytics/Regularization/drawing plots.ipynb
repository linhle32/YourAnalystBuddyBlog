```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.sort(np.random.normal(0,1,(100,1)),axis=0)
y = 2*x + np.random.normal(0,1,(100,1))
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x,y)
yh = lr.predict(x)
```

```python
lr.score(x,y)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x,y)
plt.plot(x,yh,c='red',linewidth=4)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree=10)
x10 = pf.fit_transform(x)
lr.fit(x10,y)
yh10 = lr.predict(x10)
```

```python
lr.score(x10,y)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x,y)
plt.plot(x,yh10,c='red',linewidth=4)
plt.show()
```

```python
pf = PolynomialFeatures(degree=20)
x20 = pf.fit_transform(x)
lr.fit(x20,y)
yh20 = lr.predict(x20)
```

```python
lr.score(x20,y)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x,y)
plt.plot(x,yh20,c='red',linewidth=4)
plt.show()
```
